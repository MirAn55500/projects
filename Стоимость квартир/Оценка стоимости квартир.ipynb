```python
import pandas as pd
import numpy as np
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
import xgboost as xgb

from sklearn.model_selection import GridSearchCV
from sklearn import preprocessing
from sklearn.metrics import r2_score
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
```

```python
test = pd.read_csv('Test.csv').drop('id', axis=1)
train = pd.read_csv('Train.csv').drop('id', axis=1)
```

```python
train.head()
```

```python
train = train.dropna()
```

```python
train_full = pd.get_dummies(train, columns=['date'])
test_full = pd.get_dummies(test, columns=['date'])
```

```python
train_full.head()
```

```python
scaler.fit(train_full.drop('price', axis=1))
train_scaled = scaler.transform(train_full.drop('price', axis=1))
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(train_scaled, train.price)
```

```python
train.date = str(train.date)
```

```python
from catboost import CatBoostRegressor
catboost = CatBoostRegressor(verbose=10)

catboost.fit(x_train, y_train)
y_prediction = catboost.predict(x_test) 
 
print(r2_score(y_prediction, y_test))
```

```python
from xgboost.sklearn import XGBRegressor
xgb = XGBRegressor()
xgb.fit(x_train, y_train)
y_prediction = xgb.predict(x_test)
print(r2_score(y_prediction, y_test))
```

```python
from sklearn.tree import DecisionTreeRegressor
tree = DecisionTreeRegressor()
tree.fit(x_train, y_train)
y_prediction = tree.predict(x_test)
print(r2_score(y_prediction, y_test))
```
